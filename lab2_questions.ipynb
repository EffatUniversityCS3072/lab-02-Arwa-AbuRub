#### Practice Exercise 1

Create the following NumPy arrays:
1. A 1-D array called zeros having 10 elements and all the elements are set to zero.
2. A 1-D array called vowels having the elements ‘a’, ‘e’, ‘i’, ‘o’ and ‘u’.
3. A 2-D array called ones having 2 rows and 5 columns and all the elements are set to 1 and dtype as int.
4. Use nested Python lists to create a 2-D array called myarray1 having 3 rows and 3 columns and store the following data: 2.7,  0,  -2,  3.4,  10.6,  0,  -19 99.9 13
5. A 2-D array called myarray2 using arange() having 3 rows and 5 columns with start value = 4, step size 4 and dtype as float


```python
import numpy as np

# A 1-D array called zeros having 10 elements set to zero
zeros = np.zeros(10)

# A 1-D array called vowels with elements 'a', 'e', 'i', 'o', 'u'
vowels = np.array(['a', 'e', 'i', 'o', 'u'])

# A 2-D array called ones with 2 rows and 5 columns set to 1 (int)
ones = np.ones((2, 5), dtype=int)

# A 2-D array called myarray1 with specific elements using nested lists
myarray1 = np.array([
    [2.7, 0, -2],
    [3.4, 10.6, 0],
    [-19, 99.9, 13]
])

# A 2-D array called myarray2 using arange with start=4, step=4, and float dtype
myarray2 = np.arange(4, 64, 4).reshape(3, 5).astype(float)

print("zeros:", zeros)
print("vowels:", vowels)
print("ones:\n", ones)
print("myarray1:\n", myarray1)
print("myarray2:\n", myarray2)



```

Create a 2-D array called myarray4 using arange() having 14 rows and 3 columns with start value = -1, step size 0.25 having. Split this array row wise into 3 equal parts and print the result.

```python
import numpy as np

# Create a 2-D array called myarray4 with 14 rows and 3 columns, start value -1, step size 0.25
myarray4 = np.arange(-1, -1 + 14 * 3 * 0.25, 0.25).reshape(14, 3)
print("myarray4:\n", myarray4)

# Split the array row-wise into 3 equal parts
split_arrays = np.array_split(myarray4, 3, axis=0)
print("\nSplit arrays:")
for i, arr in enumerate(split_arrays, start=1):
    print(f"Part {i}:\n", arr)



```

#### Practice Exercise 2
Read the coordinates of the capital cities of the world from the country-capital-lat-long-population.csv file. Use NumPy to print the name and coordinates of the capital city closest to the Saudi Arabia capital - Riyadh.

Note that:

1. The Country Name for Saudi Arabia is given as Saudi Arabia in the data.
2. The ‘closeness’ of capital cities from the Saudi Arabia capital is based on the Euclidean distance of their coordinates to those of the Saudi Arabia capital.


Hints:

1. Use read_html() from the Pandas library to read the table.
2. Use the to_numpy() function of the Pandas DataFrame class to convert a DataFrame to a Numpy array
3. Use broadcasting to compute the euclidean distance of capital cities from Riyadh.

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv('country-capital-lat-long-population.csv')

print(df.head())

# Calculate the closest capital city to Riyadh
from geopy.distance import great_circle

# Coordinates for Riyadh
riyadh_coordinates = (24.7136, 46.6753)

# Function to find the closest capital city
def find_closest_capital(df, riyadh_coordinates):
    closest_city = None
    closest_distance = float('inf')  # Initialize with infinity

    for index, row in df.iterrows():
        capital_coordinates = (row['Latitude'], row['Longitude'])
        distance = great_circle(riyadh_coordinates, capital_coordinates).kilometers

        if distance < closest_distance:
            closest_distance = distance
            closest_city = row['Capital Type']

    return closest_city, closest_distance

# Call the function and print the result
closest_city, closest_distance = find_closest_capital(df, riyadh_coordinates)
print(f"The closest capital city to Riyadh is {closest_city} at a distance of {closest_distance:.2f} km.")

```

#### Practical Exercise 2

Use NumPy to:

1. Print the names of the countries of the top 10 capital cities closest to the US capital - Washington DC.

2. Create and print a NumPy array containing the coordinates of the top 10 cities.

```python
from google.colab import files
import pandas as pd
from geopy.distance import great_circle
import numpy as np

# Upload the CSV file
uploaded = files.upload()

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('country-capital-lat-long-population.csv')

# Coordinates for Washington, D.C.
dc_coordinates = (38.8951, -77.0364)

# Calculate distances to Washington, D.C.
distances = []
for index, row in df.iterrows():
    capital_coordinates = (row['Latitude'], row['Longitude'])
    distance = great_circle(dc_coordinates, capital_coordinates).kilometers
    distances.append(distance)

# Convert distances to a NumPy array
distances_array = np.array(distances)

# Get the indices of the top 10 closest capitals
top_10_indices = np.argsort(distances_array)[:10]

# Get the names of the countries and coordinates for the top 10 cities
top_10_countries = df['Country'].iloc[top_10_indices].values
top_10_coordinates = df[['Latitude', 'Longitude']].iloc[top_10_indices].to_numpy()

# Print the names of the countries
print("Top 10 closest capital cities to Washington, D.C. (Countries):")
for country in top_10_countries:
    print(country)

# Print the NumPy array containing the coordinates of the top 10 cities
print("\nCoordinates of the top 10 closest capital cities:")
print(top_10_coordinates)

```
